```python
import os
import json
import shutil
from pathlib import Path
import numpy as np
from PIL import Image
import pickle
import subprocess
import glob

def create_dataset_structure(base_dir="data/instruct-pix2pix-robot"):
    """create dataset directory structure"""
    base_path = Path(base_dir)
    # create two sub-top-level directories: front and head, so that we can use them respectively if we need
    (base_path / "front").mkdir(parents=True, exist_ok=True)
    (base_path / "head").mkdir(parents=True, exist_ok=True)
    return base_dir

def save_image_pair(current_frame, future_frame, prompt_dir):
    """save image pair
    
    Args:
        current_frame: current frame image
        future_frame: future frame image
        prompt_dir: output directory
    """

    # ensure the image is PIL format
    if isinstance(current_frame, np.ndarray):
        current_frame = Image.fromarray(current_frame)
    if isinstance(future_frame, np.ndarray):
        future_frame = Image.fromarray(future_frame)

    # ensure the image is RGB mode
    if current_frame.mode != 'RGB':
        current_frame = current_frame.convert('RGB')
    if future_frame.mode != 'RGB':
        future_frame = future_frame.convert('RGB')
        
    # save image pair, and name them as 0.jpg and 1.jpg
    current_frame.save(prompt_dir / f"0.jpg")
    future_frame.save(prompt_dir / f"1.jpg")

def create_prompt_json(task_name, prompt_dir, camera_type):
    """create prompt.json file, using  two different prompts for front and head view"""
    # head
    head_prompts = {
        'block_hammer_beat': {
            'input': "A robotic arm above a table with blocks",
            'edit': "beat the block with the hammer",
            'output': "A robotic arm beating a block with a hammer"
        },
        'block_handover': {
            'input': "A robotic arm above a table with blocks",
            'edit': "handover the blocks",
            'output': "A robotic arm handing over blocks"
        },
        'blocks_stack_easy': {
            'input': "A robotic arm above a table with blocks",
            'edit': "stack blocks",
            'output': "A robotic arm stacking blocks on top of each other"
        }
    }
    
    # front
    front_prompts = {
        'block_hammer_beat': {
            'input': "A front view of a robotic workspace with blocks on the table",
            'edit': "beat the block with the hammer",
            'output': "A front view showing the robotic arm using a hammer to hit the block"
        },
        'block_handover': {
            'input': "A front view of a robotic workspace with blocks on the table",
            'edit': "handover the blocks",
            'output': "A front view showing the robotic arm lifting and handing over the blocks"
        },
        'blocks_stack_easy': {
            'input': "A front view of a robotic workspace with blocks on the table",
            'edit': "stack blocks",
            'output': "A front view showing the robotic arm stacking blocks vertically"
        }
    }
    
    prompts = front_prompts if camera_type == 'front' else head_prompts
    with open(prompt_dir / "prompt.json", "w") as f:
        json.dump(prompts[task_name], f)

def process_episodes(task_name, start_episode, end_episode, output_dir="data/instruct-pix2pix-robot"):
    """process episodes and convert data to instruct-pix2pix format"""

    base_dir = create_dataset_structure(output_dir)

    print(f"\n== Processing {task_name} episodes {start_episode} to {end_episode-1} ==\n")
    
    data_dir = f"./data/{task_name}_D435_pkl"
    task_idx = {'block_hammer_beat': 0, 'block_handover': 1, 'blocks_stack_easy': 2}[task_name]

    print(f"DEBUG : Looking for data in: {data_dir}") # todo: debug
    
    for episode_id in range(start_episode, end_episode):
        episode_dir = Path(data_dir) / f"episode{episode_id}"
        if not episode_dir.exists():
            print(f"Episode {episode_id} not found, skipping...")
            continue
            
        print(f"Processing {episode_dir}")
        
        try:
            # Choose key frames (we use frame 40 and frame 90)
            current_frame_file = episode_dir / "39.pkl"
            future_frame_file = episode_dir / "89.pkl"
            
            if current_frame_file.exists() and future_frame_file.exists():
                with open(current_frame_file, 'rb') as f:
                    current_data = pickle.load(f)
                    current_frame_front = current_data['observation']['front_camera']['rgb']
                    current_frame_head = current_data['observation']['head_camera']['rgb']
                
                with open(future_frame_file, 'rb') as f:
                    future_data = pickle.load(f)
                    future_frame_front = future_data['observation']['front_camera']['rgb']
                    future_frame_head = future_data['observation']['head_camera']['rgb']
                
                # process front view data
                front_prompt_dir = Path(base_dir) / "front" / f"prompt_{task_idx}_{episode_id}"
                front_prompt_dir.mkdir(exist_ok=True)
                save_image_pair(current_frame_front, future_frame_front, front_prompt_dir)
                create_prompt_json(task_name, front_prompt_dir, 'front')
                
                # process head view data
                head_prompt_dir = Path(base_dir) / "head" / f"prompt_{task_idx}_{episode_id}"
                head_prompt_dir.mkdir(exist_ok=True)
                save_image_pair(current_frame_head, future_frame_head, head_prompt_dir)
                create_prompt_json(task_name, head_prompt_dir, 'head')
                
                print(f"Successfully processed episode{episode_id}")
            else:
                print(f"Missing frames for episode{episode_id}")
        except Exception as e:
            print(f"Error processing episode{episode_id}: {e}")

def clean_pkl_files(task_name, start_episode, end_episode):
    """删除指定范围的pkl文件以释放空间,避免占用太多空间，导致整体生成失败"""
    
    data_dir = f"./data/{task_name}_D435_pkl"
    for episode_id in range(start_episode, end_episode):
        episode_dir = Path(data_dir) / f"episode{episode_id}"
        if episode_dir.exists():
            print(f"Cleaning up {episode_dir}")
            shutil.rmtree(episode_dir)
    
    print(f"\n== Cleaned PKL of {task_name} episodes {start_episode} to {end_episode-1} ==\n") 

def run_task_with_batching(task_name, gpu_id=0, batch_size=10, total_episodes=100):
    """run task with batching and process data after each batch"""

    # create dataset output directory
    output_dir = "data/instruct-pix2pix-robot"
    create_dataset_structure(output_dir)

    # save original config to avoid overwriting
    with open(f"./task_config/{task_name}.yml", 'r') as file:
        original_config = yaml.safe_load(file)
    
    # run generation tasks in batches
    for start_episode in range(0, total_episodes, batch_size):
        end_episode = min(start_episode + batch_size, total_episodes)
        
        # modify config file to only run this batch
        config = original_config.copy()  # load original config
        # modify config to run this batch
        config['episode_num'] = total_episodes - start_episode
        config['st_episode'] = total_episodes - end_episode
        with open(f"./task_config/{task_name}.yml", 'w') as file:
            yaml.dump(config, file, default_flow_style=False)

        print(f"DEBUG: Modified config to run episodes {total_episodes - end_episode} to {total_episodes - start_episode-1}")

        # verify config
        with open(f"./task_config/{task_name}.yml", 'r') as file:
            written_config = yaml.safe_load(file)
            print(f"\nConfig verification:")
            print(f"st_episode = {written_config.get('st_episode')}")
            print(f"episode_num = {written_config.get('episode_num')}")
                
            
        # run task
        print(f"\n======== Running {task_name} episodes {total_episodes - end_episode} to {total_episodes - start_episode-1} ========\n")
        subprocess.run(["bash", "run_task.sh", task_name, str(gpu_id)])
        
        # process and convert data
        process_episodes(task_name, total_episodes - end_episode, total_episodes - start_episode, output_dir)
        
        # clean up pkl files to save space and avoid filling up disk
        clean_pkl_files(task_name, total_episodes - end_episode, total_episodes - start_episode)
        
        print(f"\n======== Completed {task_name} episodes {total_episodes - end_episode} to {total_episodes - start_episode-1} ========\n")

def compress_dataset(output_dir="data/instruct-pix2pix-robot", output_file="robot_dataset.zip"):
    """compress dataset to a zip file"""
    print(f"Compressing dataset from {output_dir} to {output_file}...")
    subprocess.run(["zip", "-r", output_file, output_dir])
    print(f"Dataset compressed to {output_file}")

# we run the tasks in batches to avoid running out of memory
batch_size = 20
total_episodes = 100
run_task_with_batching("block_hammer_beat", gpu_id=0, batch_size=batch_size, total_episodes=total_episodes)
run_task_with_batching("block_handover", gpu_id=0, batch_size=batch_size, total_episodes=total_episodes)
run_task_with_batching("blocks_stack_easy", gpu_id=0, batch_size=batch_size, total_episodes=total_episodes)
compress_dataset()
```
